```python
import pandas as pd
import re
```

```python
df = pd.read_excel('skill_data.xlsx')
# df
```

```python
web_devloper_skills = [
    'django',
    'html',
    'css',
    'nodejs',
    'reactjs',
    'flask',
    'laravel'
]
machine_learning_skills = [
    'nosql',
    'pandas',
    'numpy',
    'matpotlib',
    'seabron',
    'scipy'
]
```

```python
skill=[]
for i in df['Skills']:
    match = re.findall(r"'(.+?)'",i)
    skill.append(list(set(match)))
```

```python
desig = []
for i in skill:
    scol = []
    w=0
    m=0
    for j in i:
        if j in web_devloper_skills:
            w+=1
        if j in machine_learning_skills:
            m+=1
    if w>=3 and m>=3:
        scol.append('Web Developer')
        scol.append('Machine Learning')
    elif w>=3:
        scol.append('Web Developer')
    elif m>=3:
        scol.append('Machine Learning')
    else:
        scol.append('Not Eligible')
    desig.append(scol)    
```

```python
df.insert(2, "Designation",desig, True)
```

```python
totalSkills=[]
for i in skill:
    totalSkills += i
```

```python
df2 = {'Name': '', 'Skills': 'Total unique skills: ' + str(len(list(set(totalSkills)))) , 'Designation': ''}
df = df.append(df2, ignore_index = True)
```
